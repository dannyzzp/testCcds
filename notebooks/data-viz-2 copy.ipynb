```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("~/projects/testCcds/data/raw/creditcard.csv")
```

```python
df
```

```python
pivotedData = df[['Amount', 'Class']].pivot_table(
    index='Class',
    values='Amount',
    aggfunc=['mean', 'std', 'count'])
```

```python
pivotedData.plot(kind='bar', y='mean', title='Average Transaction Amount by Class')
```

```python
pivotedData.plot(kind='bar', y='std', title='std of Transaction Amount by Class')
```

```python
pivotedData.plot(kind='bar', y='count', title='Number of Transaction Amount by Class')
```
